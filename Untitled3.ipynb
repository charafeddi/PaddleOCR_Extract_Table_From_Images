```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt

# Load the image from the file
img_path = r'D:\abdu\images\img1_rotated_125.jpg'
image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Perform histogram equalization using OpenCV's equalizeHist function
equalized_image = cv2.equalizeHist(image)

# Save the equalized image to a file
equalized_img_path = r'D:\abdu\images\image1_equalized.jpg'
cv2.imwrite(equalized_img_path, equalized_image)

# Convert to PIL Image format for display
original_image_pil = Image.fromarray(image)
equalized_image_pil = Image.fromarray(equalized_image)

# Display images using matplotlib for comparison
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original Image
axs[0].imshow(original_image_pil, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

# Equalized Image
axs[1].imshow(equalized_image_pil, cmap='gray')
axs[1].set_title('Equalized Image')
axs[1].axis('off')

plt.show()
```
